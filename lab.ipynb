```python
from sympy import *

x_ = symbols("x")

init_printing(use_unicode=True)

class GeoMatrix():
    def __init__(self) -> None:
        self.points = dict()

    def addPoint(self, symbol, x, y):
        self.points[symbol] = (x,y)
    
    def extend(self, a, b, symbol):
        x = b[0]+b[0]-a[0]
        y = b[1]+b[1]-a[1]
        self.addPoint(symbol, x, y)
        return x, y

    def perpandicularLine(self, k, x, y):
        b_ = symbols("b")
        kx = -(1/k)
        #bx = (sqrt(pow(x,2)+pow(y,2))/k)*sqrt(1+pow(k,2))
        bx = solve(kx*x+b_-y,b_)[0]
        return kx*x_+bx
    
    def bisectorLine(self, k1, k2):
        pass

axis = GeoMatrix()
y1 = x_+2
y2 = axis.perpandicularLine(1,-2,0)
plot(y1,y2,(x_,-5,5),aspect_ratio=(1.0,1.0))
```
